## Markers
Let us the markers of two points on the canvas. The 'o' is the marker. 

In Matplotlib's Pyplot, various markers can be used to distinguish data points on a graph. Below is a comprehensive list of marker styles and their corresponding symbols:

### **Standard Markers**:
- `'.'` : Point (small dot)  
- `','` : Pixel (single pixel, very small)  
- `'o'` : Circle  
- `'v'` : Triangle pointing downward  
- `'^'` : Triangle pointing upward  
- `'<'` : Triangle pointing left  
- `'>'` : Triangle pointing right  
- `'1'` : Tri-down (inverted triangle with tick)  
- `'2'` : Tri-up (triangle with tick)  
- `'3'` : Tri-left (left-pointing triangle with tick)  
- `'4'` : Tri-right (right-pointing triangle with tick)  
- `'8'` : Octagon  
- `'s'` : Square  
- `'p'` : Pentagon  
- `'*'` : Star  
- `'+'` : Plus  
- `'x'` : Cross (X)  
- `'D'` : Diamond  
- `'d'` : Thin diamond (rotated square)  
- `'|'` : Vertical line  
- `'_'` : Horizontal line  

---

### **Filled/Thicker Variants**:
- `'P'` : Filled plus (thicker than `'+'`)  
- `'X'` : Filled X (thicker than `'x'`)  
- `'h'` : Hexagon 1 (horizontal orientation)  
- `'H'` : Hexagon 2 (vertical orientation)  

---

### **Usage Example**:
```python
import matplotlib.pyplot as plt

plt.plot(x_data, y_data, marker='^', linestyle='None')  # Triangle-up markers
plt.plot(x_data, y_data, marker='s', linestyle='--')     # Square markers with dashed line
plt.show()
```

---

### **Notes**:
1. Marker size and color can be adjusted using `markersize`, `markerfacecolor`, and `markeredgecolor` parameters.  
   Example:  
   ```python
   plt.plot(x, y, marker='o', markersize=10, markerfacecolor='red', markeredgecolor='black')
   ```

2. To use unfilled markers, set `markerfacecolor='none'`.  
   Example:  
   ```python
   plt.plot(x, y, marker='s', markerfacecolor='none')
   ```

### Line Style
It is possible to change the default line style. Some of the line styles are solid, dotted, dashed, dashdot.

````python
x = np.arange(-25, 25)
y = 2 * x
plt.plot(x, y, c='g', linewidth = 2) # c ='g' or c ='green'
plt.plot(x, -2 * x, c='b', linestyle='dashed', linewidth = 2) # c ='b' or c ='blue'
plt.plot(np.zeros(100),  np.arange(-50, 50), c='r', linestyle='dotted', linewidth = 2) # c ='r' or c ='red'
plt.plot(x,np.zeros(50), c='y', linestyle='dashdot', linewidth = 2) # c ='y' or c ='yellow'
plt.title('X versus Y')
plt.show()
````

The alpha argument takes a value from 0 to 1. The alpha value makes the bar transparent based on the

## Adding Grid to a bar graph
The grid method add grid to the bar graph. It has color, linewidth, linestyle, alpha, axis parameter.
It is possible to add a horizontal and vertical grid using the grid method. The grid method by default provides a horizontal and vertical grid. A horizontal or vertical grid is also possible by using the axis parameter.

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha = 0.7)
plt.grid(color='gray', alpha = 0.7)
plt.show()
````

### Multiple Bar graphs on one canvas

````python
data1 = np.random.randint(0, 100, size=5)
data2 = np.random.randint(0, 100, size=5)
data3 = np.random.randint(0, 100, size=5)
width = 0.25
plt.figure(figsize=(10, 6))  # Larger figure size
plt.bar([1, 2, 3, 4, 5], data1,  width = width, color = '#7be382', edgecolor='blue',linewidth=3, label='Positive')
plt.bar(np.array([1, 2, 3, 4, 5]) + width, data2, width = width, color='#eedc82', label='Neutral')
plt.bar(np.array([1, 2, 3, 4, 5]) + width * 2, data3, width = width, color = '#cd5c5c', label='Negative')
plt.legend(loc='upper left', fontsize=12, framealpha=0.9, shadow=True)
plt.show()
````

````python
# Generate random data
np.random.seed(100)  # For reproducibility
data1 = np.random.randint(0, 100, size=5)
data2 = np.random.randint(0, 100, size=5)
data3 = np.random.randint(0, 100, size=5)
x = np.array([1, 2, 3, 4, 5])  # Category positions

# Create stacked bars
plt.bar(x, data1, color='#7be382', label='Negative', width=0.6)
plt.bar(x, data2, color='#eedc82', label='Neutral', width=0.6, bottom=data1)
plt.bar(x, data3, color='#cd5c5c', label='Positive', width=0.6, bottom=data1 + data2)

# Customize the plot
plt.title("Stacked Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(x, labels=['A', 'B', 'C', 'D', 'E'])  # Optional: Custom category labels
plt.legend(loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
````

### Normal Distribution

````python

# np.random.normal(mu, sigma, size)
# mu - mean, sigma - sigma -standard deviation, size - number of obsevations
normal_array = np.random.normal(79, 15, 10000)
normal_array
````

````python
count, bins, ignored = plt.hist(normal_array, 80, density=True)
sigma = 15
mu = 79
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
````

## Pie Chart

````python
import matplotlib.pyplot as plt

# Sample data
labels = ['Negative', 'Neutral', 'Positive']
sizes = [30, 45, 25]
colors = ['#ff9999', '#66b3ff', '#99ff99']
explode = (0.1, 0, 0)  # "Explode" the first slice

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors, 
        autopct='%1.1f%%', 
        startangle=90, 
        shadow=True)

# Add title and format
plt.title('Sample Pie Chart', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle

# Add legend
plt.legend(labels, loc="best")

plt.show()
````

## Heatmap

````python
import folium
import pandas as pd
import numpy as np

# Sample data with lat/lng
np.random.seed(42)
data = pd.DataFrame({
    'lat': np.random.uniform(40.6, 40.8, 20),  # NYC area latitudes
    'lng': np.random.uniform(-74.0, -73.9, 20),  # NYC area longitudes
    'value': np.random.randint(1, 100, 20)
})

# Create a base map
m = folium.Map(location=[40.7, -73.95], zoom_start=12, tiles='OpenStreetMap')

# Add markers
for idx, row in data.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=5,
        color='#3186cc',
        fill=True,
        fill_color='#3186cc',
        popup=f"Value: {row['value']}"
    ).add_to(m)

# Add heatmap (optional)
from folium.plugins import HeatMap
HeatMap(data[['lat', 'lng', 'value']].values.tolist()).add_to(m)

# Save or display
m.save('map.html')  # Save as HTML
m  # Display in Jupyter Notebook
````

````python
import plotly.express as px
import pandas as pd
import pycountry

# Get comprehensive country list
country_data = []
for country in pycountry.countries:
    country_data.append({
        'name': country.name,
        'iso_alpha': country.alpha_3
    })

df = pd.DataFrame(country_data)

# Generate map

fig = px.choropleth(
    df,
    locations="iso_alpha",
    color=[1]*len(df),
    hover_name="name",
    color_continuous_scale=[[0, 'lightblue'], [1, 'lightblue']],
    projection="orthographic"
)

fig.update_layout(
    title_text='World Countries (Comprehensive List)',
    coloraxis_showscale=False,
    geo=dict(showframe=False, showland=True, landcolor='white'),
     width=800,
    height=800,
)

fig.show()
````
