## Importing necessary packages

````python
import numpy as np # numerica Python, NumPy
import matplotlib.pyplot as plt
````

### Line Graphs

The plot() method draws a line from point to point on a cavas by default. The line below is between point(0.5) and (0, 10).

````python
x = np.array([0, 5])
y = np.array([0, 10])
plt.plot(x, y)
plt.show()
````

## Markers
Let us the markers of two points on the canvas. The 'o' is the marker. 

````python
x = np.array([0, 5])
y = np.array([0, 10])
plt.plot(x, y, 'o')
plt.show()
````

Let us draw a line with four points

````python
x = np.array([0, 2, 6, 8])
y  = np.array([3, 10, 3, 10])
plt.plot(x, y)
plt.show()
````

## Lines with Markers
Lines with markers allows to emphasize each data points with a certain marker style. The different markers can be found on the matplotlib [documentation](https://matplotlib.org/stable/api/markers_api.html).

````python
x = np.array([0, 2, 6, 8])
y  = np.array([3, 10, 3, 10])
plt.plot(x, y, marker ='o')
plt.show()
````

````python
x = np.array([0, 2, 6, 8])
y  = np.array([3, 10, 3, 10])
plt.plot(x, y, marker ='o')
plt.plot(x, y * 0.5, marker ='x')
plt.plot(x, y * 0.25, marker ='v')
plt.plot(x, y * 0.45, marker ='v')
plt.plot(x, y * 0.15, marker ='^')
plt.plot(x, y * 0.75, marker ='D')
plt.plot(x, y * 0.85, marker ='H')
plt.plot(x, y * 0.65, marker ='*', ms=20, mfc = 'y', mec='y')
plt.plot(x, y * 0.05, marker ='.')
plt.show()
````

### Adding color to line graph
The plot method takes many arguments. The c argument takes named color or hexa-decimal color

````python
x = np.arange(25)
y = 2 * x
plt.plot(x, y, c='g') # c ='g' or c ='green' 
plt.title('X versus Y')
plt.show()
````

### Line Width
It is possible to change the default width of the line.

````python
x = np.arange(25)
y = 2 * x
plt.plot(x, y, c='g', linewidth=5) # c ='g' or c ='green' 
plt.title('X versus Y')
plt.show()
````

### Line Style
It is possible to change the default line style. Some of the line styles are solid, dotted, dashed, dashdot. 

````python
x = np.arange(-25, 25)
y = 2 * x
plt.plot(x, y, c='g', linewidth = 2) # c ='g' or c ='green' 
plt.plot(x, -2 * x, c='b', linestyle='dashed', linewidth = 2) # c ='b' or c ='blue' 
plt.plot(np.zeros(100),  np.arange(-50, 50), c='r', linestyle='dotted', linewidth = 2) # c ='r' or c ='red' 
plt.plot(x,np.zeros(50), c='y', linestyle='dashdot', linewidth = 2) # c ='y' or c ='yellow' 
plt.title('X versus Y')
plt.show()
````

````python
x = np.arange(-25, 25)
plt.plot(x, 2 * x, c='r', linewidth = 2) # c ='r' or c ='red' 
plt.plot(x, -2 * x, c='g', linestyle='dotted', linewidth=2) # c ='g' or c ='green' 
plt.plot(x, x ** 2, c='b', linestyle='dashed', linewidth=2) # c ='b' or c ='blue'
plt.plot(x, -(x ** 2), c='y', linestyle='dashdot', linewidth=2) # c ='y' or c ='yellow' 
plt.title('X versus Y')
plt.show()
````

###  Making Scatter Plot

````python
temperature =  [
14.2,
16.4,
11.9,
15.2,
18.5,
22.1,
19.4,
25.1,
23.4,
18.1,
22.6,
17.2
]
sales = [
215,
325,
185,
332,
406,
522,
412,
614,
544,
421,
445,
408]
plt.scatter(temperature, sales)
plt.xlabel('Temperate(oC)')
plt.ylabel('Sales($)')
plt.title('Temperature versus Sales of Icecream')
plt.show()
````


## Subplots
The plt subplots() method takes three arguments that describes the layout of the figure. The layout is organized in rows and columns that are represented by the first and second argument.The third argument represents the index of the current plot

````python
plt.subplot(1, 3, 1)
plt.subplot(1, 3, 2)
plt.subplot(1, 3, 3)
plt.show()
````

````python
plt.subplot(3, 1, 1)
plt.subplot(3, 1, 2)
plt.subplot(3, 1, 3)
plt.show()
````

````python
plt.subplot(2, 2, 1)
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3)
plt.subplot(2, 2, 4)
plt.show()
````

````python
plt.subplot(2, 2, 1)
x = np.linspace(-11, 11, 100)
plt.plot(x, 2 * x)
plt.subplot(2, 2, 2)
plt.plot(x, -2 * x, c='r')
plt.subplot(2, 2, 3)
plt.plot(x, x ** 2, c='g')
plt.subplot(2, 2, 4)
plt.plot(x, x ** 3)
plt.show()
````

## Bar Graph

````python
import numpy as np # numerica Python, NumPy
import matplotlib.pyplot as plt

data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data)
plt.show()
````

## Changing the default bar color
plt.bar has a color argument and passing a value will change the color of the bars

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral')
plt.show()
````

## Adding transparncy to the bars

The alpha argument takes a value from 0 to 1. The alpha value makes the bar transparent based on the 

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha = 0.45)
plt.show()
````

## Adding Grid to a bar graph
The grid method add grid to the bar graph. It has color, linewidth, linestyle, alpha, axis parameter. 
It is possible to add a horizontal and vertical grid using the grid method. The grid method by default provides a horizontal and vertical grid. A horizontal or vertical grid is also possible by using the axis parameter.

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha = 0.7)
plt.grid(color='gray', alpha = 0.7)
plt.show()
````

### Horizontal Grid

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha=0.7)
plt.grid(color='gray', axis='y', alpha=0.7)
plt.show()
````

### Vertical Grid

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha=0.7)
plt.grid(color='gray', axis='x', alpha=0.7)
plt.show()
````

### Grid Style
The grid style could be broken lines, solid lines, dotted, stars, etc.

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha=0.7)
plt.grid(color='gray', linestyle='dashed', axis='y', alpha=0.7)
plt.show()
````

## Adding line widht to the grid

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha=0.7)
plt.grid(color='gray', linewidth = 2, linestyle='--', axis='y', alpha=0.7)
plt.show()
````

### Adding Labels to the Bar Graph

````python
data = [20, 50, 60, 80, 10]
x = [1, 2, 3, 4, 5]
plt.bar(x, data, color='coral', alpha=0.7)
plt.grid(color='gray', linewidth = 1, linestyle='--', axis='y', alpha=0.7)
plt.xlabel('X Axis')
plt.ylabel('X Axis')
plt.title('Title of the Bar graph')
plt.show()
````

### Multiple Bar graphs on one canvas

````python
data1 = np.random.randint(0, 100, size=5)
data2 = np.random.randint(0, 100, size=5)
data3 = np.random.randint(0, 100, size=5)
width = 0.25
plt.bar([1, 2, 3, 4, 5], data1,  width = width)
plt.bar(np.array([1, 2, 3, 4, 5]) + width, data2, width = width)
plt.bar(np.array([1, 2, 3, 4, 5]) + width * 2, data3, width = width, color = ['#7be382','#eedc82', '#cd5c5c'])
plt.show()
````

````python
### Stacked Bar graphs
````

````python
data1 = np.random.randint(0, 100, size=5)
data2 = np.random.randint(0, 100, size=5)
data3 = np.random.randint(0, 100, size=5)
width = 0.25
plt.bar([1, 2, 3, 4, 5], data1)
plt.bar(np.array([1, 2, 3, 4, 5]), data2)
plt.bar(np.array([1, 2, 3, 4, 5]), data3, color = ['#7be382','#eedc82', '#cd5c5c'])
plt.show()
````
